```python
Q6. List down some best practices for exception handling in python.
    ans:-
        
        Here are some best practices for exception handling in Python:

Use try-except blocks: Use try-except blocks to handle exceptions in your code. 
    This allows you to catch and handle exceptions, rather than letting them propagate up the call stack and possibly crash your program.

Be specific with exceptions: When you catch exceptions, be specific about the types of exceptions you want to catch.
    This allows you to handle specific exceptions in specific ways.

Handle exceptions gracefully: When an exception occurs, handle it gracefully.
    Provide a helpful error message to the user, log the error, and try to recover if possible.

Don't use bare except: Avoid using bare except clauses as it catches all exceptions including the unexpected ones, this makes it difficult to understand and troubleshoot issues.

Use Finally block: Use the finally block to clean up resources after an exception has occurred. This ensures that any resources that were opened or modified during the try block are properly closed or reset.

Use context managers: Context managers can help manage resources that need to be opened and closed safely. For example, use the "with" statement to open and close files automatically.

Use logging: Use Python's logging module to log exceptions and other errors. This helps with debugging and troubleshooting.

Raise exceptions: Raise exceptions when appropriate. This can help with error reporting and help the user understand what went wrong.

Follow PEP 8 conventions: Follow PEP 8 conventions for writing clean, readable, and consistent code. This includes using proper indentation, whitespace, and naming conventions.

Use third-party libraries: Consider using third-party libraries like Sentry or Rollbar to help manage exceptions and errors in your code. These libraries can help you track and analyze errors, and provide helpful debugging information.



```
